```python
from pathlib import Path
import json
import pandas as pd
```

```python
with Path("competition-corner-data.json").open() as f:
    data = json.load(f)
```

```python
workouts = pd.DataFrame(data["workouts"])
workouts.head()
```

```python
athletes = pd.DataFrame(data["athletes"])
athletes.head()
```

```python
scores = []
for athlete in data["athletes"]:
    collected = [("name", athlete["name"]), ("affiliate", athlete["affiliate"]), ("place", athlete["place"]), ("totalPoints", athlete["totalPoints"])]
    collected.extend([(workout, data["res"]) for workout, data in athlete["workoutScores"].items()])
    collected.extend([(data["workoutId"], data["points"]) for _workout, data in athlete["workoutScores"].items()])
    scores.append(collected)

print(f"{len(scores)} scores collected")
```

```python
df = pd.DataFrame([{k:v for k,v in entry} for entry in [score for score in scores]]) # , columns=["name", "affiliate", "Hoppin' and Poppin'!", "3-1-2? It's So Complex!", "Toss it, I'm Over It! (Row)", "Toss it, I'm Over It! (Metcon)", "Get on Up 'n Git Down"])
df.head()
```

```python
df = df.rename(columns={
    'workout_358': 'Hoppin and Poppin',
    'workout_359': '3-1-2 Its So Complex',
    'workout_360': 'Toss it Im Over It Row',
    'workout_361': 'Toss it Im Over It Metcon',
    'workout_362': 'Get on Up n Git Down',
    358: 'WOD1',
    359: 'WOD2',
    360: 'WOD3',
    361: 'WOD4',
    362: 'WOD5'})
df.head()
```

```python
df.to_csv("competition-corner-data.csv", index=False)
```
